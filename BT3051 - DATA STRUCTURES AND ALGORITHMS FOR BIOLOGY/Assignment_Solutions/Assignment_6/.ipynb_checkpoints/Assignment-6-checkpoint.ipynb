```python
# Defining A class Queue to Use for BFS
class Queue:
    
    def __init__ (self):
        self.queue=[]
    
    def addq (self,v):
        self.queue.append(v)
        
    def isempty(self):
        return(self.queue==[])
    
    def delq(self):
        v=None
        if not self.isempty():
            v=self.queue[0]
            self.queue=self.queue[1:]
        return(v)
    
    def __str__(self):
        return(str(self.queue))
```

```python
# This BFS Algorithm returns a Graph modelling the BFS Traversal Tree

def BFS(AList,Start_Node):
    
    global BFS_TT
    count=1
    (visited,parent) = ({},{})
    BFS_TT = nx.DiGraph()

    
    for country in AList.keys():
        visited[country] = False
        parent[country] =-1
        BFS_TT.add_node(country)
        
    q = Queue()
    visited[Start_Node] = True
    q.addq(Start_Node)
        
    while(not q.isempty()):
        country = q.delq()
        
        for child in AList[country]:
            if (not visited[child]):
                visited[child] = True
                BFS_TT.add_edge(country, child , weight=count)
                count+=1
                parent[child] = country
                q.addq(child)

    return(parent)
```

```python
# For clear display of the BFS Traversal Tree
def BFS_hierarchy_pos(G, root, width=3, vert_gap=0.1, vert_loc=0, xcenter=0.5, depth=0):
    pos = {root: (xcenter, vert_loc)}
    children = list(G.neighbors(root))

    level_width = width * (1 + depth * 1.5)
    
    if len(children) != 0:
        dx = level_width / len(children)
        nextx = xcenter - level_width / 2 - dx / 2
        for child in children:
            nextx += dx
            pos.update(BFS_hierarchy_pos(G, child, width=dx, vert_gap=vert_gap, vert_loc=vert_loc - vert_gap, 
                                         xcenter=nextx, depth=depth + 1.25))
    return pos

def Display_BFS_Traversal_Tree(Graph, Starting_Node):
    pos = BFS_hierarchy_pos(Graph, Starting_Node)

    plt.figure(figsize=(13, 6))
    nx.draw(Graph, pos=pos, with_labels=True, node_color="pink", node_size=1350, font_size=7.5, 
            font_weight="bold", arrows=True, arrowstyle='-|>')

    edge_labels = nx.get_edge_attributes(Graph, "weight") 
    nx.draw_networkx_edge_labels(Graph, pos, edge_labels=edge_labels, font_size=11, font_color="brown",
                                 bbox=dict(facecolor="white", edgecolor="none", boxstyle="round,pad=0.2"),
                                 rotate=False) 

    plt.title("Traversal Tree BFS", fontsize=15)
    plt.show()
```

```python
def DFS(AList,Starting_Node,Count=1):
    
    visited[Starting_Node] = True

    for child in AList[Starting_Node]:
        if (not visited[child]):
            parent[child] = Starting_Node
            
            DFS_TT.add_edge(Starting_Node, child, weight=Count)
            Count+=1
            
            Count=DFS(AList,child,Count)
            
    return(Count)
```

```python
class Queue:
    
    def __init__ (self):
        self.queue=[]
    
    def addq (self,v):
        self.queue.append(v)
        
    def isempty(self):
        return(self.queue==[])
    
    def delq(self):
        v=None
        if not self.isempty():
            v=self.queue[0]
            self.queue=self.queue[1:]
        return(v)
    
    def __str__(self):
        return(str(self.queue))
```

```python
def Component_Counter(AList):
    
    next_component=list(AList.keys())
    next_pointer=0
    component={}
    
    Start_Node = next_component[0]
    
    #Initalizing Component Data using cc (component counter)
    cc=1
    component[cc]=[Start_Node]
    
    #Initialization of  visited table
    visited={}
    for node in AList.keys():
        visited[node] = False
    
    #Setting up the Queue
    q = Queue()
    visited[Start_Node] = True
    q.addq(Start_Node)
    
    #completed is used to ensure once the end is reached all loops definitively end
    completed=False
    
    while (not completed):
       
        #==============================================================
        if (not q.isempty()):
            node = q.delq()
            
        #If the queue is empty and completed is False, this means we need to go to the next component of the graph
        else:
            
            #To get to the next unmarked node
            while visited[next_component[next_pointer]]==True and completed == False:
                if next_pointer == len(next_component)-1:
                    completed=True
                    
                else:
                    next_pointer+=1
                    
            if (not completed):
                #consider the next unmarked node as the new start_node
                Start_Node=next_component[next_pointer]

                cc+=1
                component[cc]=[Start_Node]
                
                visited[Start_Node] = True
                q.addq(Start_Node)
                
        #==============================================================
        for child in AList[node]:
            if (not visited[child]):
                visited[child] = True
                component[cc].append(child)
                q.addq(child)
        
    print("The number of disjoint sets/components is :",len(component))
    print("="*125)
    return(component)
```

You are responsible for planning emergency response routes for a city. The city is represented as a graph, with intersections as nodes and roads as edges.

In the event of an emergency, you need to find the shortest path from a central emergency command center to various locations in the city to minimize response time.

The city's road network is represented as an undirected weighted graph, where each road has a known travel time
(weight) associated with it.

Your task is to find the shortest paths from the central emergency command center (Node F) to all other intersections in the city.

Your code should return a nested list of [Source, Target, Duration]. Ex [[F,C,2],....]

```python
class MinHeap:
    
    # The data is stored as a tuple (node,distance) in a list
    # positions stores the node as key, and it's corresponding index in the heap as value
    def __init__(self):    
        self.heap = []
        self.positions = {}

    # Get the index of the parent, follow 0 based index
    def parent(self, i):
        return (i - 1) // 2

    # Get the index of the left child
    # Note this can also return values out of index as such, we shall always check if left < len(self.heap)
    def left_child(self, i):
        return (2 * i) + 1

    # Get the index of the right child
    # Note this can also return values out of index as such, we shall always check if right < len(self.heap)
    def right_child(self, i):
        return (2 * i) + 2
    
    # =================================================================

    # Heapify down from index i to maintain the min-heap property
    def heapify_down(self, i):
        smallest = i
        left = self.left_child(i)
        right = self.right_child(i)
        n = len(self.heap)

        # Find the smallest of the current node and its children
        if left < n and right < n:
            
            if self.heap[left][1] < self.heap[smallest][1] and self.heap[left][1] <= self.heap[right][1]:
                smallest = left
            elif right < n and self.heap[right][1] < self.heap[smallest][1] and self.heap[right][1] < self.heap[left][1]:
                smallest = right
                
        elif left < n and right >= n:
             if self.heap[left][1] < self.heap[smallest][1]:
                    smallest = left
                    
        # If the smallest is not the current node, swap and continue heapifying
        if smallest != i:
            self.positions[self.heap[i][0]],self.positions[self.heap[smallest][0]] = self.positions[self.heap[smallest][0]],self.positions[self.heap[i][0]]
            self.heap[i], self.heap[smallest] = self.heap[smallest], self.heap[i]
            self.heapify_down(smallest)
            
        # Breaks out once it is done heapifying recursively
        else:
            return
    
    # =================================================================

    # Heapify the entire array (build-heap) in O(n) time
    def heapify(self):
        n = len(self.heap)
        
        # O(n) Initialization of the Vertex to Heap Map
        for k in range(n):
            self.positions[self.heap[k][0]] = k
        
        # Start heapifying from the last non-leaf node downwards, also O(n)
        for i in range(((n - 1)//2), -1, -1):
            self.heapify_down(i)
            
    # =================================================================
    
    # Pops the top-most element and returns the tuple as the output
    def extract_min(self):
        if len(self.heap) == 0:
            return None
        
        root = self.heap[0]
        self.positions[self.heap[-1][0]] = 0
        del (self.positions[self.heap[0][0]])

        self.heap[0] = self.heap[-1]
        self.heap.pop()
        
        if len(self.heap) > 0:
            self.heapify_down(0)
            
        return(root)

    # Return the minimum element (root) of the heap
    def get_min(self):
        if len(self.heap) > 0:
            return self.heap[0] 
        else:
            return(None)

    # =================================================================
    
    # Heapify up to restore the heap property after updation
    def heapify_up(self, i):
        while i > 0:
            if self.heap[self.parent(i)][1] > self.heap[i][1]:
                self.positions[self.heap[i][0]] , self.positions[self.heap[self.parent(i)][0]] = self.positions[self.heap[self.parent(i)][0]] , self.positions[self.heap[i][0]]
                self.heap[i], self.heap[self.parent(i)] = self.heap[self.parent(i)], self.heap[i]
                i = self.parent(i)
            else:
                i=0
    
    # =================================================================
    
    # Given a node, update the data and ensure it satisfies the heap property post update
    def update(self,node,value):
        
        # Updating value
        ind = self.positions[node]
        self.heap[ind] = (self.heap[ind][0] , value)
        
        parent = self.parent(ind)
        left = self.left_child(ind)
        right = self.right_child(ind)
        n = len(self.heap)
        
        LC = (left < n )
        RC = (right < n)
        P = (parent >= 0)
        
        #Restoring Heap property
        if P == True :
            self.heapify_up(ind)
                
        if LC == True and RC == True :
            self.heapify_down(ind)
        
        elif LC == True and RC == False:
            if self.heap[left][1] < self.heap[ind][1]:
                self.positions[self.heap[i][0]],self.positions[self.heap[ind][0]] = self.positions[self.heap[ind][0]],self.positions[self.heap[i][0]]
                self.heap[i], self.heap[ind] = self.heap[ind], self.heap[i]
        
        else:
            return
    # =================================================================
```

```python
def Shortest_Path(AList,source):
                                      
    (visited,distance) = ({},{})
    inp=[]
    
    # Initalization of visited and distance tables
    for node in AList.keys():
        if node != source :
            (visited[node],distance[node]) = (False,float('inf'))
            inp.append((node, float('inf')))
        else:
            (visited[node],distance[node]) = (False, 0)
            inp.append((node, 0))
    
    # Initalizing the Min Heap
    dist_heap=MinHeap()
    dist_heap.heap=inp
    dist_heap.heapify()
    
    for n in range(len(AList)):
        
        # This is O(log n)
        min_dist = dist_heap.extract_min()
        next_dist = min_dist[1]
        next_node = min_dist[0]
        
        # Disconnected portions of the graph cannot be reached, as such their distance shall remain infinite
        if next_dist == float('inf'):
            break
                                   
        visited[next_node] = True
        
        
        for (node,d) in AList[next_node]:
            if not visited[node]:
                distance[node] = min(distance[node],distance[next_node]+d)
                dist_heap.update(node,distance[node])
    
    shortest_path=[]
    
    for node in sorted(AList.keys()):
        shortest_path.append([source,node,distance[node]])
            
    return(shortest_path)
```

“My whole life is a lie” - True

Output :“My name is Lance Ambu”, “My name is Bart”    

```python
class SuffixTrie:
    
    def __init__(self):
        self.root = {}
                
    # =======================================================================================
    # Takes a single string as input and creates branches in the suffix trie           
    def update (self,text):
        
        # The * marks the start of a text line, the dollar marks the end. 
        # The property of the * is used in the search function below.
        text = "* " + text[:-1] + " $"
        
        # Returns a list of all the words in the text
        seq = text.split(" ")
        #print(seq)
        # n - start index, m - iterative index
        for n in range(len(seq)):
            current_node = self.root
            
            for m in seq[n:]:
                
                if m == "$":
                    current_node[m] = None
                
                elif m not in current_node.keys():
                    current_node[m] = {}
                
                current_node = current_node[m]
    # =======================================================================================
    # A function to check if a given sentence was in the inital data set.
    # Note the sentences considered here are without punctuation
    
    def search (self,sentence):
        current_node = self.root
        mod_s = "* " + sentence + " $"
        seq = mod_s.split(" ")
        
        for word in seq:
            if word not in current_node.keys():
                print (False)
                return
    
            current_node = current_node[word]
        
        print (True)
        return
    
    # =======================================================================================
    # Function to read all posiible paths from a given start_node
    # start_node is the starting pointing, text denotes the list aldready visited strings.
    def read (self , start_node):
        text=[]
        
        for n in start_node.keys():
            temp=[]
            
            if n != "$":
                s = n + " " #Bart 
                next_node = start_node[n]
                temp = self.read(next_node) #[""] 
                
                for k in range(len(temp)):
                    if temp[k] != "":
                        temp[k] = s + temp[k]
                    else:
                        temp[k] = s.rstrip(" ")
            else:
                temp.append("")
        
        
            for sequence in temp:
                text.append(sequence)
                    
        return(text)
    
    # =======================================================================================
    def completion (self,text):
        
        current_node = self.root
        seq = text.split(" ")
        autocomplete=""
        results=[]
        
        # The loop gets to the end of the text, current_node waits on the last word of text
        for word in seq:
            if word not in current_node.keys():
                print ("No matches")
                return
            
            autocomplete = autocomplete + word+" "
            current_node = current_node[word]
        
        data=self.read(current_node)
        
        for sentence in data:
            results.append(autocomplete + sentence )
        
        pprint.pprint(results)
        print("="*100)
        return
    # =======================================================================================
```

**<u>Question 5</u>** 

**a)** Use a single regular expression to tokenize the given SMILES (Simplified Molecular Input Line Entry System) strings representing chemical compounds into atoms, bonds and other symbols.

Terms within square brackets need to be kept as is. The SMILES strings and their expected tokens are given

**b)** Given a SMILES string, write a function to check if it is a valid one. The rules here are:

● The following are valid symbols for atoms: C, N, O, F, P, S, Cl, Br, I, and their
lowercase equivalents.

● Bonds can be described by =, #, and .

● Branches represented by () should be balanced (every ‘(‘ should have a matching
‘)’)

● [] is allowed. Any other character apart from the given ones are not allowed.

```python
import re

# ========================================================================================================
def tokenize_smiles(smiles):
    pattern = r'Cl|Br|[CNOFPSI]|[=#\.]|[\(\)]|\[[^\[^\]]*\]|\d'
    
    tokens = re.findall(pattern, smiles, re.IGNORECASE)
    
    return(tokens)
# ========================================================================================================
def SMILES_isvalid(smiles):
    tokens=tokenize_smiles(smiles)
    braces=[]
    reconstruct=""
    result=True
    search_pattern= r'[^A-Z=#\.\(\)\d\+\-@]'
    
    for token in tokens:
        if token == "(":
            braces.append(token)
            
        elif token == ")":
            if braces != []:
                braces.pop()
            else:
                result=False
                return(result)
        
        if token.isdigit():
            result=False
            return(result)
        
        if token[0]== "[":
            n=len(token)
            seq=token[1:n-1]
            if bool(re.search(search_pattern, seq, re.IGNORECASE)):
                result=False
                return(result)

        reconstruct+=token
    
    if reconstruct != smiles:
        result=False
    
    return(result)
# ========================================================================================================
```

```python
smiles_list = [
    "N[C@@H](C)C(=O)O",
    "N#N",
    "[Cu+2].[O-]S(=O)(=O)[O-]",
    "O1C=C[C@H]",
    "[O@@H]CO-[PHI]",
    "[COO-@]C[@NH3+]"
]

for smiles in smiles_list:
    tokens = tokenize_smiles(smiles)
    print("SMILES :",smiles)
    print("Tokens:",tokens)
    print("Is the sequence valid :",SMILES_isvalid(smiles))
    print("="*100)
```

```python
data=[]
with open(r"Q6_updated.txt",'r') as f1:
    text=f1.readlines()
    for three_mer in text:
        if three_mer[0] in ["A","T","C","G"]:
            seq=three_mer.rstrip("\n")
            data.append(seq)
            
print(data)
```

```python
import pprint
import networkx as nx
import matplotlib.pyplot as plt

def Genome_sequence(data):
    De_Brujin = nx.MultiDiGraph() #Creating a De-Brujin Graph
    edges=[]
    suffix=[]
    prefix=[]
    
    #=================================================================================================================
    # Constructing the De-Brujin Graph
    for seq in data:
        suffix.append(seq[1:])
        prefix.append(seq[:2])
        edges.append(seq)
            
        De_Brujin.add_edge(seq[:2],seq[1:], weight=seq)
    
    #=================================================================================================================
    # Displaying the De-Brujin Graph
    plt.figure(figsize=(10, 9))  

    # Setting the background color
    plt.gca().set_facecolor('gold')

    pos = nx.circular_layout(De_Brujin)
    nx.draw_networkx_nodes(
        De_Brujin, pos, node_size=450, node_color='skyblue', edgecolors='black', linewidths=0.75)
    
    nx.draw_networkx_labels(De_Brujin, pos, font_size=12)

    nx.draw_networkx_edges(
        De_Brujin, pos,
        connectionstyle="arc3,rad=0.115", 
        arrows=True  
    )

    plt.title("De-Brujin Graph", size=16)
    
    plt.show()
    
    #=================================================================================================================
    
    # Determining if an Eulerian Circuit is possible for the given data
    suffix.sort()
    prefix.sort()
    possible=True

    if suffix == prefix:
        possible=True
    else:
        possible=False
    
    genome=""
    
    if possible == True:
        path=list(nx.eulerian_circuit(De_Brujin))
        genome+=path[0][0]
        for info in path:
            genome+=info[1][1]
        
        print("The Genome Sequence is :",genome[:-2])
        print("="*100)
        print("The number of nodes in the graph :", len(list(De_Brujin.nodes)))
        print("="*100)
        print("The total number of edges in the graph :",len(data))
        
        return(genome[:-2])
        
        
    else:
        print("No Eulerian Circuit exists in the given sequence")
        return(None)
        print("="*100)
```

```python
genome=Genome_sequence(data)
```

As is to be expected of a De-Brujin Graph :

The number of nodes is $4^{k-1}$, as more the number of kmers given it is more likely that all possible (k-1)mers will be present in the graph (for smaller k values at least, like k=3 here).

The number of edges in the graph represents the number of input k-mers, this is because each edge links a suffix and prefix of a k-mer and as such the number of edges must be the number of input k-mers.

The number of nodes in the graph is much lesser than the number of kmer reads given.  
The number of edges is equal to the number of kmer reads given.

Since Eulerian Circuits are not unique, it is also possible for different traversal methods to give different genome sequences as the output.

**NOTE**   
The edges have their corresponding weights as the 3-mers.  
They could not be displayed due to space limitations (It leads to an undecipherable graph).  
Thus these labels have been left out for clarity's sake.

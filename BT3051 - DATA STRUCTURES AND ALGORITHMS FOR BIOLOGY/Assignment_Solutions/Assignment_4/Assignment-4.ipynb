In genomics, researchers often need to assemble DNA fragments into a full genome. Each fragment represents a portion of the genome, and some fragments overlap with others. The goal is to reconstruct the genome with the minimum number of fragments. Each fragment has a start position and an end position.

You are given a list of DNA fragments, where each fragment is represented by its start and end positions. Write a function to find the minimum number of fragments required to cover the entire genome from position 1 to n using a greedy approach. 

<u>**Function:**</u>   
```
def min_dna_fragments(fragments, n):
    
    Output: Returns the minimum number of DNA fragments required to cover the genome.
    return: int, minimum number of fragments required to cover the genome.
    
    Parameters:  
    fragments: List of tuples (start, end), where each tuple represents the start and end positions of a fragment.  
    n: int, the length of the genome.  
   
```

<u>**Sample**</u>    
fragments = [(1, 4), (3, 5), (0, 6), (5, 10), (8, 12)]    
n = 12  
min_dna_fragments(fragments, n)  
 
Output Example:   
3 # Minimum number of fragments (e.g., fragments [(0, 6), (5, 10), (8, 12)])  

=====================================================================================================================

In the data, the segments (start,end), both are considered inclusive.

Sample:  
fragments=[(0, 3), (4, 6), (4, 9)]     
n=9  

The output returned will be 2.  
#The sequence will be [(0,3),(4,9)]

=====================================================================================================================

````python
def min_dna_fragments(fragments, n):

    x = sorted(fragments)
    seq=[]    
    count=0
    
    #Appending the first element into the sequence
    for start in range(len(x)):
        
        if x[0][0]!=0:
            print("Incomplete Sequence, Hole in between.")
            return(-1)
            
        elif start > 0 and x[start][0] != 0:
            seq += [x[start-1]]
            count = start
            break
        
        elif start==len(x)-1:
            
            if x[start][1]==n:
                seq += [x[start]]
                count=len(x)
                
            else:
                print("Incomplete Sequence, Hole in between.")
                return(-1)
            
    #A variable used as a buffer to ensure that only the most optimum value gets pushed into the list
    #temp ensures only those values whose start is in range of the end of the last term in sequence and whose end spans the longest range get appended.
    temp=[]
    
    while count < len(x):
        
        if x[count][1]>seq[-1][1]:

            if temp==[]:
                if x[count][0]>(seq[-1][1]+1):
                    print("Incomplete Sequence, Hole in between.")
                    print("="*100)
                    return(-1)  
                else:
                    temp=[x[count]]
                
            elif temp[0][1]<x[count][1]:
                temp=[x[count]]
        
        if count<len(x)-1:
            if x[count+1][0]>(seq[-1][1]+1):
                seq+=temp
                temp=[]
                    
        else:
            seq+=temp
            del(temp)
            
        count+=1
    #-------------------------------------------- End of While Loop --------------------------------------------
    
    if seq[-1][1]==n:
        print("Complete Sequence.")
        print(seq)
        print("="*100)
        return(len(seq))
        
    else:
        print("Incomplete Sequence, Total Sequence cannot be obtained from given data.")
        print("n=",n)
        print("="*100)
        return(-1)
````

In gene expression studies, researchers need to activate certain genes for specific time intervals during an experiment. Each gene activation represents a critical experiment interval where the gene is expressed. However, due to resource limitations (e.g., using a specific tool to express genes), only one gene can be activated at a time.

Given a list of gene expression intervals, each with a start and end time, your task is to select the maximum number of non-overlapping gene expression intervals to run during the experiment. Write a greedy algorithmic (python code) to solve the Interval Scheduling Problem.    
Also Explain the Approach used in the code.
   
**Input:**
A list of n gene expression intervals. Each interval is represented as a tuple (start_time, end_time) where start_time is the time when the gene starts expressing, and end_time is the time when the gene stops expressing.  
**Output:**  
Return the maximum number of non-overlapping gene expression intervals that can be scheduled.


<u>**Sample**</u>  
Input:  
intervals = [(1, 3), (2, 5), (4, 7), (6, 8)]

Output:  
Maximum number of non-overlapping intervals: 2    
Explanation : you can select the intervals (1, 3) and (6, 8) as they do not overlap. This allows for two gene expression activations without conflict.

=====================================================================================================================
Since the question doesn't make mention of what happens in the case where start and stop times overlap, it is assumed in the case they are equal, both the sequences are allowed to be considered. 

Example:  
interval4 = [(1, 3), (2, 4), (3, 5), (4, 6), (5, 7)]  

Output:  
[(1, 3), (3, 5), (5, 7)]  

=====================================================================================================================

````python
def max_genes(data):
    print("The input data is:",data)
    #print("="*100)
    
    schedule=[]
    temp=[]
    for k in data:
        temp.append(k[::-1])
        
    temp.sort()
    #temp now contains the list in (stop,start) order and is sorted in ascending order of end times
    
    schedule.append(temp[0][::-1])
    
    for n in range(1,len(temp)):
        if temp[n][1]>=schedule[-1][1]:
            schedule.append(temp[n][::-1])
    
    del(temp)
    print("The optimum scheduling as per requirement is :",schedule)
    
    return(len(schedule))
````

In genetics, palindromic sequences play a crucial role in biological processes like DNA replication and transcription. A palindromic DNA sequence is one that reads the same forward and backward. Given a DNA sequence, your task is to partition the sequence such that every subsequence in the partition is a palindromic DNA sequence.

Write a python program to return all possible ways to partition the DNA sequence where every partition is a palindrome using Dynamic Programming .  
Also Explain the Approach
   
Input:   
A string dna_sequence, consisting of nucleotides ('A', 'T', 'C', 'G').

Output:  
Return a list of lists, where each inner list represents a valid partition of palindromic subsequences.

<u>**Sample**</u>  
Input:  
dna_sequence = "ATTA"

Output:  
[   
["A", "T", "T", "A"],  
["A", "TT", "A"],  
["ATTA"]  
]


````python
def mapping(sequence):
    
    n = len(sequence)
    map = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(i,n):
            subseq=sequence[i:j+1]
            if subseq == subseq[::-1]:
                map[i][j]=1
    
    return(map)

def splits(sequence):
    
    map = mapping(sequence)
    n=len(sequence)
    
    temp = {}
    temp[0] = [[]]
    
    for i in range(n):
        for j in range(i,n):
            
            x = temp.get(i)
            
            if map[i][j]==1:
                
                for subseq in x:
                    
                    next = subseq + [sequence[i:j+1]]
                    
                    if temp.get(j+1) is None:
                        temp[j+1] = [next]
                    
                    else:
                        temp[j+1].append(next)
                        
    return(temp[n])  
````

````python
gene="CTTCAACATTAAACGGTGAACCTACAGACGTGTA"
op=splits(gene)
for seq in op:
    print(seq)
````

In RNA folding, predicting secondary structure involves finding the maximum number of non-overlapping base pairs that form between certain bases (e.g., A-U and G-C). Given an RNA sequence, the task is to find the maximum number of base pairs that can be formed, using dynamic programming.

Write a dynamic programming solution to calculate the maximum number of valid base pairs that can be formed in an RNA sequence, ensuring no overlapping pairs.

<u>**Function:**</u>  
```
def max_rna_pairs(rna_sequence):

    Finds the maximum number of valid base pairs that can form in an RNA sequence.
    
    Parameters: 
    rna_sequence: str, the RNA sequence
    
    Ouptut:  
    return: int, the maximum number of valid base pairs

pass
```

<u>**Sample**</u>    
Input:  
max_rna_pairs("AUGCUAGC")

Output Example:   
3 # Maximum number of base pairs that can be formed (e.g., A-U, G-C, G-C)

````python
#Consideration 1 as stated in the documentation

import numpy as np

def max_rna_pairs1(seq):
    
    n = len(seq)

    map = np.zeros((n,n))
    pairs = [["A","U"],["U","A"],["G","C"],["C","G"]]

    
    for i in range(n):
        map[i,i] = 0  # A pair cannot be 1 nucleotide long, upper of the table is only of concern

    
    for diff in range(1, n):
        for i in range(n-1, diff-1, -1):
            j = i - diff
            tmax = 0
            
            
            for k in range(j, i):
                tmax = max(tmax, map[j, k] + map[k+1, i])
            
            
            if [seq[i], seq[j]] in pairs:
                map[j, i] = max(tmax, 1 + map[j+1, i-1])  
            else:
                map[j, i] = tmax 
                
                
    print(seq)            
    #print(map)
    print("Maximum number of pairs:", map[0, n-1])
    print("="*100)
````

````python
#Consideration 2 as stated in the documentation
def max_rna_pairs2(seq):
    data={}
    data["A"],data["U"],data["G"],data["C"]=0,0,0,0
    temp=seq.upper()
    for n in temp:
        if n=="A":
            data["A"]+=1
        if n=="U":
            data["U"]+=1
        if n=="G":
            data["G"]+=1
        if n=="C":
            data["C"]+=1
            
    maxlen=min(data["A"],data["U"])+min(data["G"],data["C"])
    return(maxlen)
````

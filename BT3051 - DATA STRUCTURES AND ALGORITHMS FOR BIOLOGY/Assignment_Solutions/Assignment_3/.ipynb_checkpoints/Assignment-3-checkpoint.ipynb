(7 marks)**

A pharmaceutical company has many drugs in its production pipeline and will select a drug for large scale production based on two criteria: market demand (high demand = high priority) and production cost (high production cost = low priority).

The criteria can change with time.The company’s R&D team works on new drugs which can be added to the production pipeline. Choose a suitable data structure to represent the drugs in the production pipeline along with their market demand and production cost.

Implement this as a class where you can add drugs to the production pipeline, change the priority of the company, and select a drug for production. Simulate a company’s one year production plan (Jan - Dec) based on the following conditions:

$1.$ At the start of the year, the company had 10 drugs on the production list. In January, the production criteria was market demand. A drug was selected based on its market demand score.

$2.$ Till April, the same drug was continued while 4 more drugs were added (1 in thebeginning of each month) to the production list. In May, the company changed its criteria to production cost and an appropriate drug was selected from the list and was continued till the end of the year while new drugs were added to the list every month.

$3.$ Print the drug with the highest priority at the end of each month (not the one in production)

<u>Note:</u>   
I. Generate random numbers between 0 and 100 for the drug properties (market demand and production cost).  
II. You would have added 22 drugs in total at the end of the year.    

```python
# Defining a class Heap to use for the problem
class heap:

    # lc-left child, rc-right child, pp - parent pointer
    def __init__(self, data=None, mode="MD"):

        self.data = data
        self.mode = mode

        self.pp = None
        self.lc = None
        self.rc = None
        self.count = 0

    def parent(self, pos):
        # Returns the current position where the next append is to take place.
        # Every ith element is the parent of the 2i+1 and 2i+2 elements in a heap.
        ind = (pos - 1) // 2
        return ind

    def terminal(self):
        pathway = []
        parent = self.parent(self.count + 1)
        pathway.append(parent)

        while parent > 0:
            parent = self.parent(parent)
            pathway.append(parent)

        end = self

        for ind in pathway[-2::-1]:
            if ind % 2 != 0:
                end = end.lc
            else:
                end = end.rc

        return end

    def insert(self, data, mode="MD"):

        if self.data is None:
            self.data = data
            self.mode = mode

        else:
            end = self.terminal()
            mode = end.mode
            node = heap(data, mode)
            self.count += 1

            if (self.count % 2) != 0:
                end.lc = node
                node.pp = end
            else:
                end.rc = node
                node.pp = end

            adjust = node

            if adjust.mode == "MD":
                while adjust.data[1] > adjust.pp.data[1] and adjust.pp.pp is not None:
                    adjust.data, adjust.pp.data = adjust.pp.data, adjust.data
                    adjust = adjust.pp

                if adjust.data[1] > adjust.pp.data[1]:
                    adjust.data, adjust.pp.data = adjust.pp.data, adjust.data
            else:
                while adjust.data[2] < adjust.pp.data[2] and adjust.pp.pp is not None:
                    adjust.data, adjust.pp.data = adjust.pp.data, adjust.data
                    adjust = adjust.pp

                if adjust.data[2] < adjust.pp.data[2]:
                    adjust.data, adjust.pp.data = adjust.pp.data, adjust.data

    # Input must be a list
    def heapify(self, database, mode="MD"):
        for data in database:
            self.insert(data, mode)

        return self.count

    def drop_terminal(self):

        if self.lc is None and self.data is not None:
            data = self.data
            self.data = None

        elif self.data is None:
            return "Empty Heap"

        else:
            pathway = []
            parent = self.parent(self.count)
            pathway.append(parent)

            while parent > 0:
                parent = self.parent(parent)
                pathway.append(parent)

            end = self

            for ind in pathway[-2::-1]:
                if ind % 2 != 0:
                    end = end.lc
                else:
                    end = end.rc

            if end.rc is None:
                data = end.lc.data
                end.lc.pp = None
                end.lc = None
            else:
                data = end.rc.data
                end.rc.pp = None
                end.rc = None
            self.count -= 1

        return data

    def breakdown(self):
        database = []
        recur = self.count + 1
        while recur > 0:
            x = self.drop_terminal()
            database.append(x)

            recur -= 1

        delete = self.drop_terminal()

        return database

    # Define change priority

    def priorchange(self, newmode):
        if newmode not in ["MD", "PC"]:
            return

        temp = self.breakdown()
        self.heapify(
            temp,
            newmode,
        )
        del temp
```

**<u>Problem 2:</u>  
(10 marks)**

Two companies PiedPiper and Hooli are in their hiring process. They receive 10,000 resumes from the same set of applicants. Both the companies evaluate the resumes and score each of them between 0 to 1000. They sort the resumes from highest to lowest and add them to their database after which they call the top 100 applicants for the interview process.The person with the highest resume score will be the first one to be interviewed.

Both companies use different strategies to sort the resumes:  
a) PiedPiper chooses a random resume and compares the others with that in a recursive fashion.   
b) Hooli divides the resumes into two halves to sort them recursively.  

Simulate this interview process and check which company goes through the interview process quickly and explain why.

Note: What is expected: Two sorting algorithms for resume sorting and a data structure to represent the database and how applicants are called for the interview. Generate random numbers between 0 to 1000 for resume scores.

```python
import random

def Piedpiper(L,l=0,r=None): # Sort L[l:r]
    
    if r==None:
        r=len(L)
    
    if (r - l <= 1):
        return(L)
    
    pivotind=random.randint(l,r-1)

    L[l],L[pivotind]=L[pivotind],L[l]
    pivot,lesser,greater= L[l],l+1,l+1
    
    for i in range(l+1,r):
        
        if L[i] < pivot:
            L[lesser],L[greater]=L[i],L[lesser]
            lesser+=1
            greater+=1
        else:
            greater+=1

    L[l],L[lesser-1]=L[lesser-1],L[l]
    
    Piedpiper(L,l,lesser-1)
    
    Piedpiper(L,lesser,greater)
    
    return(L)
```

```python
def merge(l1,l2):
    sort=[]
    while l1!=[] and l2!=[]:
        if l1[0]>l2[0]:
            sort.append(l2.pop(0))
        else:
            sort.append(l1.pop(0))
    
    if l1==[]:
        sort.extend(l2)
    else:
        sort.extend(l1)
        
    return(sort)

def Hooli(l):
    
    if len(l)<=1:
        return(l)
    
    
    section=len(l)//2
    
    lhalf=Hooli(l[:section])
    rhalf=Hooli(l[section:])
    
    mergesorted=merge(lhalf,rhalf)
    
    return(mergesorted)
```

```python
import time
class timer:
    def __init__(self):
        self.start_time=0
        self.elapsed_time=0
        
    def start(self):
        self.start_time=time.perf_counter()
        
    def stop(self):
        self.elapsed_time=time.perf_counter()-self.start_time
        return(self.elapsed_time)
```

```python
Hooliavg=0
Piedpiperavg=0
swH=timer()
swPP=timer()
for simulations in range(500):
    test=[]
    for n in range(10000):
        test.append(random.randint(0,1000))
    
    swH.start()
    Hooli(test)
    Hooliavg+=swH.stop()
    
    swPP.start()
    Piedpiper(test)
    Piedpiperavg+=swPP.stop()
    
Hooliavg=Hooliavg/500
Piedpiperavg=Piedpiperavg/500
print("The average time taken by Hooli's system to sort is:",Hooliavg)
print("The average time taken by Piedpiper's system to sort is:",Piedpiperavg)
```

You are designing a digital note-taking system where you can add, update, retrieve and delete notes (heading and content). In this system, unique identifiers are generated from the note heading. Implement this using python.

Note: Python dictionaries are not allowed. You can use built-in hash functions.

```python
class Hashmap:

    def __init__(self):
        self.heading = None
        self.value = [None] * 10000

    # Function to add notes -> write
    def write(self, heading, text):

        key = hash(heading)
        ind = key % 10000

        if self.value[ind] is None:
            self.value[ind] = [[key, text]]
        else:
            self.value[ind].append([key, text])

    # Update lets you rewrite the notes in a given heading
    def update(self, heading):

        key = hash(heading)
        ind = key % 10000

        # if the index is empty
        if self.value[ind] is None:

            print("Invalid Heading, Not in Database")
            print("=" * 100)

            return

        # if the index exists
        else:
            for iter in range(len(self.value[ind])):

                if self.value[ind][iter][0] == key:

                    print("The current stored data is:")
                    print(heading, "\n", self.value[ind][iter][1])
                    notes = ""

                    control = True
                    while control:

                        s = input("Enter the new notes:")
                        notes += s

                        x = input(
                            "Enter F to finish updating, any other key to continue:"
                        )
                        if x.upper() == "F":
                            control = False

                    self.value[ind][iter][1] = notes
                    print("=" * 100)
                    return

        # In the case of collisions
        print("Invalid Heading, Not in Database")
        print("=" * 100)
        return

    # The append function, only adds data to the end of the current notes
    def append(self, heading):

        key = hash(heading)
        ind = key % 10000

        # if the index is empty
        if self.value[ind] is None:

            print("Invalid Heading, Not in Database")
            print("=" * 100)

            return

        # if the index exists
        else:
            for iter in range(len(self.value[ind])):

                if self.value[ind][iter][0] == key:

                    notes = self.value[ind][iter][1]

                    control = True
                    while control:

                        s = input("Enter the new notes:")
                        notes += s

                        x = input(
                            "Enter F to finish appending, any other key to continue:"
                        )
                        if x.upper() == "F":
                            control = False

                    self.value[ind][iter][1] = notes
                    print("=" * 100)
                    return

        # In the case of collisions
        print("Invalid Heading, Not in Database")
        print("=" * 100)
        return

    # Fucntion to retreive notes
    def read(self, heading):

        key = hash(heading)
        ind = key % 10000

        # if the index is empty
        if self.value[ind] is None:

            print("Invalid Heading, Not in Database")
            print("=" * 100)

            return

        # if the index exists
        else:
            for data in self.value[ind]:

                if data[0] == key:

                    print("\033[4m" + heading + "\033[0m", "\n", data[1])
                    print("=" * 100)
                    return

        # In the case of collisions
        print("Invalid Heading, Not in Database")
        print("=" * 100)
        return

    def delete(self, heading):

        key = hash(heading)
        ind = key % 10000

        # if the index is empty
        if self.value[ind] is None:

            print("Invalid Heading, Not in Database")
            print("=" * 100)
            return

        # if the index exists
        else:
            for data in self.value[ind]:

                if data[0] == key:
                    print("\033[4m" + heading + "\033[0m", "\n", data[1])

                    if len(self.value[ind]) == 1:
                        self.value[ind] = None
                    else:
                        self.value[ind].remove(data)

                    print("\n The above data has been deleted.")
                    print("=" * 100)
                    return

        # In the case of collisions
        print("Invalid Heading, Not in Database")
        print("=" * 100)
        return
```

Provide examples for adding and finding proteins.

```python
class Tree:

    #info=[protein ID,protein name, protein function]
    #lc-left child, rc-right child
    
    def __init__(self,info=None):
        self.info = info
        self.lc = None
        self.rc = None
        
        self.height=0
    
    # Only empty node has info None
    def isempty(self):
        return (self.info is None)
    
    # Leaf nodes have data and both children empty
    def isleaf(self):
        return ((self.info is not None) and (self.lc is None) and (self.rc is None))
    
    
    def inorder(self):
        
        if self.isempty():
            return([])
        
        else:
            #if self.lc.isempty():
                #print(self.info,"YES")
            #if self.rc.isempty():
                #print(self.info,"YES")
            
            return(self.lc.inorder()+[self.info]+self.rc.inorder())
        
    # Display Tree as a list
    def __str__(self):
        information=""
        x=self.inorder()
        for protein in x:
            s=str(protein)+"\n"
            information+=s
        information=information.rstrip("\n")
        return(information)
    
    
    # Check if info v occurs in tree
    #This functions returns the data of the element if it is present, else it returns None and prints Invalid ID.
    def find(self,ID):
        if self.isempty():
            print("="*125)
            print("Invalid ID, Not in Database.")
            print("="*125)
            return
        
        if self.info[0] == ID:
            print("="*125)
            print("Protein name:",self.info[1])
            print("Protein Function:",self.info[2])
            print("="*125)
            return
        
        if ID < self.info[0]:
            return(self.lc.find(ID))
        
        if ID > self.info[0]:
            return(self.rc.find(ID))
    

    def LLrotate(self):
        rightinfo=self.info
        leftinfo=self.lc.lc.info
        maininfo=self.lc.info
        
        rrt=self.rc
        rlt=self.lc.rc
        llt=self.lc.lc.lc
        lrt=self.lc.lc.rc
        
        lefthead=Tree(leftinfo)
        righthead=Tree(rightinfo)
        
        self.info=maininfo
        self.height=2
        
        self.rc=righthead
        self.rc.height=1
        self.lc=lefthead
        self.lc.height=1
        
        self.lc.lc=llt
        self.lc.rc=lrt
        self.rc.rc=rrt
        self.rc.lc=rlt

        
        return
    
    def RRrotate(self):
        rightinfo=self.rc.rc.info
        leftinfo=self.info
        maininfo=self.rc.info
        
        llt=self.lc
        lrt=self.rc.lc
        rrt=self.rc.rc.rc
        rlt=self.rc.rc.lc
        
        lefthead=Tree(leftinfo)
        righthead=Tree(rightinfo)
        
        self.info=maininfo
        self.height=2
        
        self.rc=righthead
        self.rc.height=1
        self.lc=lefthead
        self.lc.height=1
        
        self.lc.lc=llt
        self.lc.rc=lrt
        self.rc.rc=rrt
        self.rc.lc=rlt
        
        return
    
    def LRrotate(self):
        rightinfo=self.info
        leftinfo=self.lc.info
        maininfo=self.lc.rc.info
        
        rrt=self.rc
        rlt=self.lc.rc.rc
        lrt=self.lc.rc.lc
        llt=self.lc.lc
        
        lefthead=Tree(leftinfo)
        righthead=Tree(rightinfo)
        
        self.info=maininfo
        self.height=2
        
        self.rc=righthead
        self.rc.height=1
        self.lc=lefthead
        self.lc.height=1
        
        self.lc.lc=llt
        self.lc.rc=lrt
        self.rc.rc=rrt
        self.rc.lc=rlt
        
        return
    
    def RLrotate(self):
        rightinfo=self.rc.info
        leftinfo=self.info
        maininfo=self.rc.lc.info
        
        llt=self.lc
        lrt=self.rc.lc.lc
        rlt=self.rc.lc.rc
        rrt=self.rc.rc
        
        lefthead=Tree(leftinfo)
        righthead=Tree(rightinfo)
        
        self.info=maininfo
        self.height=2
        
        self.rc=righthead
        self.rc.height=1
        self.lc=lefthead
        self.lc.height=1
        
        self.lc.lc=llt
        self.lc.rc=lrt
        self.rc.rc=rrt
        self.rc.lc=rlt
        
        return
    
    def slope(self):
        if self.info is None:
            return(0)

        slope=self.lc.height-self.rc.height
        return(slope)

    
    def rebalance(self):
        
        bf=self.slope()
        
        if bf in [-1,0,1]:
            #print("Did nothing")
            #print("="*100)
            return
        else:
            #print("Tryna reorder")
            
            if bf==2 and self.lc.slope()==1:
                self.LLrotate()
                
            elif bf==2 and self.lc.slope()==-1:
                self.LRrotate()
                
            elif bf==-2 and self.rc.slope()==1:
                self.RLrotate()
                
            elif bf==-2 and self.rc.slope()==-1:
                self.RRrotate()
            
            #print("="*100)
            
    def insert(self,data):
        if self.isempty():
            self.info = data
            self.height=1
            self.lc = Tree()
            self.rc = Tree()
            
        if self.info[0] == data[0]:
            return
        
        if data[0] < self.info[0]:
            self.lc.insert(data)
            self.height=1+max(self.lc.height,self.rc.height)
            self.rebalance()
            return
        
        if data[0] > self.info[0]:
            self.rc.insert(data)
            self.height=1+max(self.lc.height,self.rc.height)
            self.rebalance()
            return
```

Use appropriate data structure to represent this; add 1000 people (identified by unique IDs 1 to 1000) to the queue and remove them when they are leaving.

```python
class Stack:
    
    def __init__(self):
        self.info=[]
        
    def push(self,ID):
        self.info.append(ID)
        return
    
    def dump(self,data):
        for ID in data:
            self.push(ID)
            
    def population(self):
        return(len(self.info))
            
    def pop(self):
        return(self.info.pop())
        
    def breakdown(self):
        for ID in self.info[-1::-1]:
            print(ID)
        return
```

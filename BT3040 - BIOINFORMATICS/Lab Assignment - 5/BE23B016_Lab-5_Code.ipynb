```python
# Question 5
def matching_pentapeptides(seq1,seq2):

  penta_1 = {}
  match_locations = {}

  # Setting up the initial match strings, and determining their frequency in the first string
  for res in range(len(seq1)-4):

    if penta_1.get(seq1[res:res+5]) is None:
      penta_1[seq1[res:res+5]] = 1

    else:
      penta_1[seq1[res:res+5]] += 1

  # Getting the pentapeptides that occur in both sequences and the frequency of the subsequence in each string
  for res in range(len(seq2)-4):

    if penta_1.get(seq2[res:res+5]) is not None:

      if match_locations.get(seq2[res:res+5]) is None:
        match_locations[seq2[res:res+5]] = (penta_1.get(seq2[res:res+5]), 1)

      else:
        match_locations[seq2[res:res+5]] = (
            match_locations.get(seq2[res:res+5])[0],
            match_locations.get(seq2[res:res+5])[1]+1
            )

  for subseq in match_locations.keys():
    print(f"The pentatpeptide {subseq} occurs {match_locations[subseq][0]} time(s) in Sequence 1 and {match_locations[subseq][1]} time(s) in Sequence 2")
```

```python
# Question 6
# NOTE: Alignment symbols obtained from BLAST's pairwise sequence alignment is used as the input here to calculate alignment parameters
def alignment_paramaters(align_seq, query_seq):

  identity, similarity, gaps, mismatches = 0, 0, 0, 0
  coverage = (len(align_seq)/len(query_seq)) * 100

  for res in align_seq:

    if res.isalpha():
      identity += 1
      similarity += 1

    elif res == "+":
      similarity += 1

    elif res == "-":
      gaps += 1

    else:
      mismatches += 1

  scores={}
  scores["identity"] = (identity/len(align_seq))*100
  scores["similarity"] = (similarity/len(align_seq))*100
  scores["gaps"] = (gaps/len(align_seq))*100
  scores["mismatches"] = (mismatches/len(align_seq))*100

  print(f"The query coverage is  {coverage} %")
  print(f"The sequence identity is {identity} / {len(align_seq)} or {round(scores["identity"],3)}%")
  print(f"The sequence similarity is {similarity} / {len(align_seq)} or {round(scores["similarity"],3)}%")
  print(f"The gap occurence is {gaps} / {len(align_seq)} or {round(scores["gaps"],3)}%")
  print(f"The number of mismatches between the sequences is: {mismatches}; {round(scores["mismatches"],3)}% of the sequence")
```

```python
# Question 1
def calc_composition(seq):

  comp = {}
  AAs = "ACDEFGHIKLMNPQRSTVWY"

  for AA in AAs:
    comp[AA] = False

  for residue in seq:
    if comp.get(residue) is False:
      comp[residue] = 1
    else:
      comp[residue] += 1

  for residue in comp.keys():
    comp[residue] = (comp[residue] * 100) / len(seq)

  return(comp)
```

```python
comp1 = calc_composition(seq1)
for AA in comp1.keys():
  print(f"The residue {AA} has a composition of: {round(comp1[AA],2)} %")
print("="*125)

comp2 = calc_composition(seq2)
for AA in comp2.keys():
  print(f"The residue {AA} has a composition of: {round(comp2[AA],2)} %")
print("="*125)

comp3 = calc_composition(seq3)
for AA in comp3.keys():
  print(f"The residue {AA} has a composition of: {round(comp3[AA],2)} %")
print("="*125)
```

```python
# Question 2

def Sequence_Weight(seq):

  AA_W = {
      'A': 85,   # Ala
      'C': 115,  # Cys
      'D': 130,  # Asp
      'E': 145,  # Glu
      'F': 160,  # Phe
      'G': 70,   # Gly
      'W': 200,  # Trp
      'H': 150,  # His
      'I': 125,  # Ile
      'K': 145,  # Lys
      'L': 125,  # Leu
      'M': 143,  # Met
      'N': 130,  # Asn
      'Y': 175,  # Tyr
      'P': 110,  # Pro
      'Q': 140,  # Gln
      'R': 170,  # Arg
      'S': 100,  # Ser
      'T': 115,  # Thr
      'V': 110   # Val
      }
  weight = 0

  for residue in seq:
    weight += AA_W[residue]

  weight -= ((len(seq)-1)*18)

  print(f"The molecular weight of the given sequence is: {weight}")
  print("="*100)
```

```python
Sequence_Weight(seq1)
Sequence_Weight(seq2)
Sequence_Weight(seq3)
```

```python
# Question 3

def similarity(seq):
  Group_data = {
    'A': (8.47, 8.95),  # Ala
    'D': (5.97, 5.91),  # Asp
    'C': (1.39, 0.47),  # Cys
    'E': (6.32, 4.78),  # Glu
    'T': (5.79, 6.54),  # Thr
    'F': (3.91, 3.68),  # Phe
    'G': (7.82, 8.54),  # Gly
    'H': (2.26, 1.25),  # His
    'I': (5.71, 4.77),  # Ile
    'V': (7.02, 6.76),  # Val
    'K': (5.76, 4.93),  # Lys
    'L': (8.48, 8.78),  # Leu
    'M': (2.21, 1.56),  # Met
    'N': (4.54, 5.74),  # Asn
    'W': (1.44, 1.24),  # Trp
    'P': (4.63, 3.74),  # Pro
    'Q': (3.82, 4.75),  # Gln
    'R': (4.93, 5.24),  # Arg
    'S': (5.94, 8.05),  # Ser
    'Y': (3.58, 4.13)   # Tyr
  }

  comp = calc_composition(seq)
  scores_1, scores_2 = [], []

  sum_sc_1, sum_sc_2 = 0, 0

  AAs = "ACDEFGHIKLMNPQRSTVWY"
  for AA in AAs:
    scores_1 = abs(comp[AA]-Group_data[AA][0])
    scores_2 = abs(comp[AA]-Group_data[AA][1])
    sum_sc_1 += scores_1
    sum_sc_2 += scores_2

    if scores_1 > scores_2:
      print(f"Given sequence's {AA} is similar to Group A")
    elif scores_2 > scores_1:
      print(f"Given sequence's {AA} is similar to Group B")
    else:
      print(f"Given sequence's {AA} is similar to Group A and Group B")

  # print(sum_sc_1,sum_sc_2)
  if sum_sc_1 < sum_sc_2:
    print(f"Given sequence is similar to Group A")
  elif sum_sc_1 > sum_sc_2:
    print(f"Given sequence is similar to Group B")
  else:
    print(f"Given sequence is similar to both Groups")

  print("="*100)
```

```python
similarity(seq1)
similarity(seq2)
similarity(seq3)
```

```python
# Question 4
def count_di(seq):
  Scores_1 = np.zeros((20, 20))
  Scores_2 = np.zeros((20, 20))
  Scores_3 = np.zeros((20, 20))
  Score_list_1, Score_list_2, Score_list_3 = [], [], []

  AAs = "ACDEFGHIKLMNPQRSTVWY"
  di_nuc = {}

  for n in range(len(seq)-1):
    di = seq[n:n+2]
    if di_nuc.get(di) is None:
      di_nuc[di] = 1
    else:
      di_nuc[di] += 1

  comp = {}
  for AA in AAs:
    comp[AA] = 0

  for residue in seq:
    comp[residue] += 1


  for n in range(20):
    for m in range(20):
      if di_nuc.get(AAs[n]+AAs[m]) is not None:
        Scores_1[n][m] = (di_nuc[AAs[n]+AAs[m]]*100)/(comp[AAs[n]] + comp[AAs[m]])
        Score_list_1.append((Scores_1[n][m], AAs[n]+AAs[m]))

        Scores_2[n][m] = (di_nuc[AAs[n]+AAs[m]]*100)/(len(seq)-1)
        Score_list_2.append((Scores_2[n][m], AAs[n]+AAs[m]))

        Scores_3[n][m] = (di_nuc[AAs[n]+AAs[m]]*100)/(comp[AAs[n]] * comp[AAs[m]])
        Score_list_3.append((Scores_3[n][m], AAs[n]+AAs[m]))

  pd.set_option('display.max_columns', None)
  pd.set_option('display.width', None)
  pd.set_option('display.expand_frame_repr', False)

  aa = list("ACDEFGHIKLMNPQRSTVWY")
  df1 = pd.DataFrame(Scores_1, index=aa, columns=aa)
  print(df1.round(2))
  Score_list_1.sort(reverse = True)
  for k in range(min(10,len(Score_list_1))):
    print(f"{k+1}) The residue pair: {Score_list_1[k][1]}, with an occurrence of {round(Score_list_1[k][0],3)} %")
  print("-"*100)

  df2 = pd.DataFrame(Scores_2, index=aa, columns=aa)
  print(df2.round(2))
  Score_list_2.sort(reverse = True)
  for k in range(min(10,len(Score_list_2))):
    print(f"{k+1}) The residue pair: {Score_list_2[k][1]}, with an occurrence of {round(Score_list_2[k][0], 3)} %")
  print("-"*100)

  df3 = pd.DataFrame(Scores_3, index=aa, columns=aa)
  print(df3.round(2))
  Score_list_3.sort(reverse = True)
  for k in range(min(10,len(Score_list_3))):
    print(f"{k+1}) The residue pair: {Score_list_3[k][1]}, with an occurrence of {round(Score_list_3[k][0], 3)} %")
  print("="*100)
```

```python
count_di(seq1)
count_di(seq2)
count_di(seq3)
```

```python
Et = "1.90   1.52   2.04   1.54   1.86   1.90   1.76   1.95   1.37   1.97   1.96   1.56   1.70   1.52   1.48   1.75   1.77   1.98   1.87   1.69"
Ca = "20.00  26.00  25.00  33.00  46.00  13.00  37.00  39.00  46.00  35.00  43.00  28.00  22.00  36.00  55.00  20.00  28.00  33.00  61.00  46.00"
Hgm = "13.85  11.61  15.37  11.38  13.93  13.34  13.82  15.28  11.58  14.13  13.86  13.02  12.35  12.61  13.10  13.39  12.70  14.56  15.48  13.88"

AAs = "ADCEFGHIKLMNPQRSTVWY"

Et_temp = Et.split("   ")
Et = {}

Ca_temp = Ca.split("  ")
Ca = {}

Hgm_temp = Hgm.split("  ")
Hgm = {}

for n in range(len(AAs)):
  Et[AAs[n]] = float(Et_temp[n])
  Ca[AAs[n]] = float(Ca_temp[n])
  Hgm[AAs[n]] = float(Hgm_temp[n])

def Scoring(seq):
  score_Hgm, score_Ca, score_Et = 0, 0, 0
  for AA in seq:
    score_Hgm += Hgm[AA]
    score_Ca += Ca[AA]
    score_Et += Et[AA]

  avg_Hgm = score_Hgm/len(seq)

  print(f"Average hydrophobicity for the given sequence is: {round(avg_Hgm, 2)}")
  print(f"Helical contact area for the given sequence is: {score_Ca}")
  print(f"Total non-bonded energy for the given sequence is: {round(score_Et, 3)}")
  print("="*100)
```

```python
Scoring(seq1)
Scoring(seq2)
Scoring(seq3)
```

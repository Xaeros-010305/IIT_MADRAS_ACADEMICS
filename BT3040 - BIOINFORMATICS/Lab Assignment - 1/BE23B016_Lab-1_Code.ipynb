```python
# Question 3
def find_complement(seq):

    map = {'A': 'T', 'T': 'A', 'G': 'C', 'C': 'G'}

    seq = seq.upper()
    comp_seq = ""

    for nt in seq:
        comp_seq += map[nt]

    comp_seq = comp_seq[-1::-1]

    return(comp_seq)

seq = "CTCGGATTTGTAAAGATCATGATCTCATACATAGTACCTAGCCATTG"
complement = find_complement(seq)
print(seq)
print(complement)
```

```python
# Question 5
def translation(seq):

    codon_table = {
        'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M',
        'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T',
        'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K',
        'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',
        'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L',
        'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P',
        'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q',
        'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R',
        'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V',
        'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A',
        'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E',
        'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G',
        'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S',
        'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L',
        'TAC':'Y', 'TAT':'Y', 'TAA':'*', 'TAG':'*',
        'TGC':'C', 'TGT':'C', 'TGA':'*', 'TGG':'W',
        'CT' : 'L', 'GT' : 'V', 'TC' : 'S', 'CC' :'P', 'AC' : 'T', 'GC' : 'A', 'GC' : 'R', 'GG' : 'G'}

    translations = {}
    seq1,seq2,seq3 = "", "", ""

    for nt in range (0,len(seq),3):

        # Forward Frame 1
        if nt+2 < len(seq):
            seq1 += codon_table[seq[nt:nt+3]]

        elif nt+1 < len(seq):

            if codon_table.get(seq[nt:nt+2]):
                seq1 += codon_table[seq[nt:nt+2]]
            else:
                print(seq[nt+2:nt+4])
                seq1 += "X"
        else:
            seq1 += "X"

        # Forward Frame 2
        if nt+3 < len(seq):
            seq2 += codon_table[seq[nt+1:nt+4]]

        elif nt+2 < len(seq):

            if codon_table.get(seq[nt+1:nt+3]):
                seq2 += codon_table[seq[nt+1:nt+3]]
            else:
                print(seq[nt+1:nt+3])
                seq2 += "X"

        else:
            seq2 += "X"

        # Forward Frame 3
        if nt+4 < len(seq):
            seq3 += codon_table[seq[nt+2:nt+5]]

        elif nt+3 < len(seq):

            if codon_table.get(seq[nt+2:nt+4]):
                seq3 += codon_table[seq[nt+2:nt+4]]
            else:
                print(seq[nt+2:nt+4])
                seq3 += "X"

        else:
            seq3 += "X"

    translations[1] = seq1
    translations[2] = seq2
    translations[3] = seq3

    seq = find_complement(seq)
    seq1,seq2,seq3 = "", "", ""

    for nt in range (0,len(seq),3):

        # Reverse Frame 1
        if nt+2 < len(seq):
            seq1 += codon_table[seq[nt:nt+3]]

        elif nt+1 < len(seq):

            if codon_table.get(seq[nt:nt+2]):
                seq1 += codon_table[seq[nt:nt+2]]
            else:
                print(seq[nt+2:nt+4])
                seq1 += "X"
        else:
            seq1 += "X"

        # Reverse Frame 2
        if nt+3 < len(seq):
            seq2 += codon_table[seq[nt+1:nt+4]]

        elif nt+2 < len(seq):

            if codon_table.get(seq[nt+1:nt+3]):
                seq2 += codon_table[seq[nt+1:nt+3]]
            else:
                print(seq[nt+1:nt+3])
                seq2 += "X"

        else:
            seq2 += "X"

        # Reverse Frame 3
        if nt+4 < len(seq):
            seq3 += codon_table[seq[nt+2:nt+5]]

        elif nt+3 < len(seq):

            if codon_table.get(seq[nt+2:nt+4]):
                seq3 += codon_table[seq[nt+2:nt+4]]
            else:
                print(seq[nt+2:nt+4])
                seq3 += "X"

        else:
            seq3 += "X"

    translations[4] = seq1
    translations[5] = seq2
    translations[6] = seq3

    return(translations)
```

```python
# Question 6
# Assuming first occurence and no overlap
def search_1(seq,subseq):

    locations, start = [], 0

    while seq[start:].find(subseq) != -1:
        ind = (seq[start:].find(subseq))
        locations.append(str(start+ind))
        start += ind + len(subseq)

    print("Total match(es):", len(locations))
    print(f"Position(s) of match:{', ' .join(locations)}")
    return

# Assuming overlap is allowed
def search_2(seq,subseq):

    locations, start = [], 0

    while seq[start:].find(subseq) != -1:
        ind = (seq[start:].find(subseq))
        locations.append(str(start+ind))
        start += ind + 1

    print("Total match(es):", len(locations))
    print(f"Position(s) of match:{', ' .join(locations)}")
    return

seq = "GACATTGTGAACAGTAAAAAAGTCCATGCAATGCGCAAGGAGCAGAAGAGGAAGCAGGGCAAGCAGCGCTCCATGGGCTCTCCCATGGACTACTCTCCTCTGCCCATCGACAAGCATGAGCCTGAATTTGGTCCATGCAGAAGAAAACTGGATGGG"
subseq = input("Enter the sequence to be searched: ").upper()
search_1(seq,subseq)
print("="*125)
search_2(seq,subseq)
```

```python
# Showing the difference between the 2 methods
search_1("ATATAT", "ATAT") # A nucleotide can be a part of only 1 sequence
print("="*125)
search_2("ATATAT", "ATAT") # A nucleotide can be a part of more than 1 sequence
```

```python
# Question 8 (bse - base stacking energy)
def average_bse(seq):

    bse_table = {
        "AA": -4, "AT": -7, "AC": -5, "AG": -11,
        "TA": -7, "TT": -2, "TC": -3, "TG": -4,
        "CA": -9, "CT": -5, "CC": -6, "CG": -7,
        "GA": -9, "GT": -6, "GC": -4, "GG": -11
    }

    bse = 0

    for n in range(len(seq)-1):
        bse+=bse_table.get(seq[n:n+2])

    abse = bse/(len(seq)-1)

    return(abse)
```

```python
seq = "CTCGGATTTGTAAAGATCATGATCTCATACATAGTACCTAGCCATTG"
print(average_bse(seq))
```

```python
def GC_content(seq):
    GC = 0
    for nt in seq:
        if nt in ("G","C"):
            GC+=1

    GCp = (GC/len(seq))*100
    return(GCp)

print(GC_content("AAATGGCCCTAA"))
print("="*125)
print(GC_content("TTAGGGCCATTT"))
print("="*125)
```
